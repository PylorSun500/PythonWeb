## 蓝图 `blueprint`

蓝图，类似于多个HTML管理一套网页的功能。<br>
每个蓝图表示一个子页面，由一个主页面来管理。<br>
如果按照结构来看，也就是：<br>
Flask 实例作为 kernel ，来管理一堆 Blueprint 实例。

蓝图工作流包括以下流程：
1. 创建蓝图子页面

    [e.g. 用户页面](./blueprint/user.py)

    如代码所示，蓝图子页面需要实例化的并非 `Flask` ，而是 `Blueprint`，<br>
    作为子页面，蓝图也没有形如 `app.run()` 的启动函数。

2. 创建主文档

    [e.g. index.py](./blueprint/index.py)

    如代码所示，主页面创建思路类似于一般页面。<br>
    但需要在导入库后 **导入目录下的子页面** 。<br>
    还需要使用
    ```python
    app.register_blueprint()
    ```
    来将子页面注册到主文档中。
