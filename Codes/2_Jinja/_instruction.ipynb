# Jinja 2 框架

Jinja 2 资产分为两部分：**Python 脚本** 与 **Jinja 模板** 。<br>
Python 脚本用以调用模板以及管理变量；
Jinja 模板存放于 ./templates 目录中。本质上是一些 HTML 文件，用以定义即将展示的网页样式。<br>

Flask 提供 `render_template` 渲染 HTML 页面。

后者负责渲染 HTML 。<br>

在脚本中定义变量，并在 `return` 中将变量传入 HTML 。

## Jinja 模板

Jinja 模板本质上是 Jinja 引用的 HTML 。作为一种 HTML 文件，大多数语法标签都与普通 HTML 类似，但有些内容具有区别。

### 变量

Jinja 可以引用变量，这些变量来自 Python 。当需要添加 Python 语句和表达式时：<br>

表达式由 `{{ ... }}` 定界；<br>
语句由 `{% ... %}` 定界；<br>
注释由 `{# ... #}` 定界。<br>

但要注意，在 Jinja 中，HTML 注释 `<!-- -->` 不再可用。
